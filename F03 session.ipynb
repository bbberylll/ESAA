## 내용 정리

1. 기본 단계
    - data 로딩 --> 머신러닝 알고리즘 모델 선택 --> 데이터 세트 학습/테스트로 분리
    - 데이터 :  dataframe 형태로 바꿔줘야 함
          .data  .target  .target_names   .feature_names
      
    - 학습용 데이터와 테스트용 데이터를 분리해야 하는 이유
      = 모델의 성능을 정확하게 판단하기 위해서
      --> model_selection 모듈에서 train_test_split() import
  
    - train_test_split() :
      1. X_train = 학습용 피처 데이터
      2. X_test = 테스트용 피처 데이터
      3. y_train = 학습용 레이블 데이터 세트
      4. y_test = 테스트용 레이블 데이터 세트
     
    - 데이터 기반의 학습과 예측 + 예측 성능 평가 수행
      1. 사용할 머신러닝 알고리즘의 객체 생성
      2. fit(X_train, y_train) : 학습 수행 
      3. predict(X_test) : 예측 수행
      4. accuracy_score(y_teat, 3번의 예측값) : 정확도 기반의 예측 성능 평가


2. 사이킷런 기반의 프레임 워크
   1) fit, predict() 메서드
      - 지도 학습 : Classification과 Regression 클래스(=Estimator 클래스)에 모두 포함돼 있음
      - 비지도 학습 : fit(), transform() 사용하지만 의미는 다름. fit()의 경우 학습이 아닌 데이터 형태에 맞춰 데이터의 구조를 맞추는 작업 수행
                    fit_transform()이라는 메서드도 있음
  
   2) 사이킷런의 주요 모듈
      - 피처 처리 : sklearn.preprocessing / sklearn.decomposition  / sklearn.model_selection
      - 평가 : sklearn.metrics
      - 알고리즘 : sklearn.ensemble  / sklearn.linear_model  / sklearn.tree  / sklearn.cluster  

3. Model Selection 모듈 소개
   1) 학습/테스트 데이터 세트 분리 by train_test_split()
      - 전체 데이터로 fit하고 predict하면 정확도가 100%가 나오게 됨  --> 모델의 성능을 파악할 수 없음
      - 전체 데이터에서 test_size 파라미터를 이용해 학습/테스트 데이터를 분리해주게 된다!
        
   2) 교차 검증  : 데이터의 갯수가 적은 경우 학습/테스트로 분리하게 되면 모델의 성능이 더 떨어지게 된다. 그래서 나오게 된 과적합을 방지하면서도 모델의 성능을 떨어뜨리지 않을 수 있는 방법!!
      - K fold 교차 검증 : K개의 데이터 폴드 세트를 만들어 K번 만큼 학습/테스트 역할을 바꿔가면서 성능 평가
      - Stratified K폴드 : 불균형한 분포를 가진 레이블 데이터일 경우 적용하는 교차 검증 방법
      - cross_val_score() : 교차 검증을 한번에 수행해주는 API == 분류일 경우 Stratified로, 회귀일 경우 K fold 방법으로 검증 수행 
     
    3) GridSearchCV : 교차 검증과 최적 하이퍼 파라미터 튜닝을 한 번에 수행 

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

import pandas as pd

iris = load_iris()

iris_data = iris.data

iris_label = iris.target
print('iris target값:', iris_label)
print('iris target명:', iris.target_names)

iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_df['label'] = iris.target
iris_df.head(3)

X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.2, random_state=11)
```

```python
dt_clf = DecisionTreeClassifier(random_state=11)

dt_clf.fit(X_train, y_train)
pred = dt_clf.predict(X_test)

from sklearn.metrics import accuracy_score
print('예측 정확도: {0:.4f}'.format(accuracy_score(y_test,pred)))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np

iris = load_iris()
features = iris.data
label = iris.target
dt_clf = DecisionTreeClassifier(random_state=156)

kfold = KFold(n_splits=5)
cv_accuracy = []
print('붓꽃 데이터 세트 크기:',features.shape[0])
n_iter = 0

for train_index, test_index  in kfold.split(features):
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]
    dt_clf.fit(X_train , y_train)    
    pred = dt_clf.predict(X_test)
    n_iter += 1
    accuracy = np.round(accuracy_score(y_test,pred), 4)
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]
    print('\n#{0} 교차 검증 정확도 :{1}, 학습 데이터 크기: {2}, 검증 데이터 크기: {3}'
          .format(n_iter, accuracy, train_size, test_size))
    print('#{0} 검증 세트 인덱스:{1}'.format(n_iter,test_index))
    cv_accuracy.append(accuracy)
    
print('\n## 평균 검증 정확도:', np.mean(cv_accuracy)) 
```

```python
import pandas as pd

iris = load_iris()

iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['label']=iris.target
iris_df['label'].value_counts()
kfold = KFold(n_splits=3)

n_iter =0
for train_index, test_index  in kfold.split(iris_df):
    n_iter += 1
    label_train= iris_df['label'].iloc[train_index]
    label_test= iris_df['label'].iloc[test_index]
    print('## 교차 검증: {0}'.format(n_iter))
    print('학습 레이블 데이터 분포:\n', label_train.value_counts())
    print('검증 레이블 데이터 분포:\n', label_test.value_counts())
    
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=3)
n_iter=0

for train_index, test_index in skf.split(iris_df, iris_df['label']):
    n_iter += 1
    label_train= iris_df['label'].iloc[train_index]
    label_test= iris_df['label'].iloc[test_index]
    print('## 교차 검증: {0}'.format(n_iter))
    print('학습 레이블 데이터 분포:\n', label_train.value_counts())
    print('검증 레이블 데이터 분포:\n', label_test.value_counts())
dt_clf = DecisionTreeClassifier(random_state=156)

skfold = StratifiedKFold(n_splits=3)
n_iter=0
cv_accuracy=[]

for train_index, test_index  in skfold.split(features, label):
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]
    #학습 및 예측 
    dt_clf.fit(X_train , y_train)    
    pred = dt_clf.predict(X_test)

    n_iter += 1
    accuracy = np.round(accuracy_score(y_test,pred), 4)
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]
    print('\n#{0} 교차 검증 정확도 :{1}, 학습 데이터 크기: {2}, 검증 데이터 크기: {3}'
          .format(n_iter, accuracy, train_size, test_size))
    print('#{0} 검증 세트 인덱스:{1}'.format(n_iter,test_index))
    cv_accuracy.append(accuracy)
    
print('\n## 교차 검증별 정확도:', np.round(cv_accuracy, 4))
print('## 평균 검증 정확도:', np.round(np.mean(cv_accuracy), 4))
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, 
                                                    test_size=0.2, random_state=121)
dtree = DecisionTreeClassifier()

parameters = {'max_depth':[1,2,3], 'min_samples_split':[2,3]}
import pandas as pd

grid_dtree = GridSearchCV(dtree, param_grid=parameters, cv=3, refit=True)

grid_dtree.fit(X_train, y_train)

scores_df = pd.DataFrame(grid_dtree.cv_results_)
scores_df[['params', 'mean_test_score', 'rank_test_score', \
           'split0_test_score', 'split1_test_score', 'split2_test_score']]
print('GridSearchCV 최적 파라미터:', grid_dtree.best_params_)
print('GridSearchCV 최고 정확도: {0:.4f}'.format(grid_dtree.best_score_))
estimator = grid_dtree.best_estimator_

pred = estimator.predict(X_test)
print('테스트 데이터 세트 정확도: {0:.4f}'.format(accuracy_score(y_test,pred)))
```
